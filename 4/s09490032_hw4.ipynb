```python
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup

#將網頁資料利用requests套件GET下來，儲存在response變數中
url = 'https://udn.com/news/breaknews/1'

response = requests.get(url) 

html = response.text

#解析HTML
data = BeautifulSoup(html, 'html.parser')
```

```python
#尋找HTML中所有標籤為"section"且屬性class為"cate-list__subheader"的元素
titles = data.find_all("section", class_= "cate-list__subheader")

print(titles)
```

```python
#定義categories串列(每個項目都是一個元組，包含了分類的名稱、路徑和索引)
categories = [
    ('精選', '/news/breaknews/1/0#breaknews', 0),
    ('要聞', '/news/breaknews/1/1#breaknews', 1),
    ('娛樂', '/news/breaknews/1/8#breaknews', 8),
    ('財經', '/news/breaknews/1/6#breaknews', 6),
    ('文教', '/news/breaknews/1/12#breaknews', 12)
]

#使用for迴圈對categories進行迭代。在每次迭代中，從元組中提取出分類名稱、路徑和索引
for category, path, index in categories:
    
    #使用 data.select_one() 方法來尋找符合路徑條件的元素，並將結果存儲在 link 變數中
    link = data.select_one('a[href^="{}"]'.format(path))
    
    #如果找到了符合條件的元素，則從該元素中獲取 href 屬性的值，並使用正則表達式從中提取出分類的 ID
    if link:
        
        href = link['href']
        
        category_id = re.search(r'/(\d+)/{}#breaknews'.format(index), href).group(1)
        
        #根據分類的 ID 和索引構建完整的分類連結 category_link
        category_link = 'https://udn.com/news/breaknews/{}/{}#breaknews'.format(category_id, index)
        
        print(href)
        
        print(category_link)
```

```python
categories = [
    ('精選', 0),
    ('要聞', 1),
    ('娛樂', 8),
    ('財經', 6),
    ('文教', 12)
]

#使用for迴圈對categories進行迭代。在每次迭代中，從元組中提取出分類名稱和索引
for category, index in categories:
    
    #根據索引構建分類的路徑 path，使用該路徑和基本網址 'https://udn.com' 構建完整的 URL url
    path = '/news/breaknews/1/{}#breaknews'.format(index)
    
    url = 'https://udn.com' + path
    
    response = requests.get(url)
    
    html = response.text

    data = BeautifulSoup(html, 'html.parser')

    #使用 data.find_all() 方法尋找所有包含新聞標題的元素，並將結果存儲在 Title 變數中
    Title = data.find_all("h2")

    #定義一個空串列filtered_titles用於存儲符合條件的新聞標題
    filtered_titles = []

    
    #使用for迴圈抓取Title中的每個元素，同時追蹤索引
    for index, title in enumerate(Title, start=1):
        
        Title_text = title.text.strip()
        
        #當標題不以'{{'開頭且不包含冒號時，才將其添加到filtered_titles中
        if not Title_text.startswith('{{') and ':' not in Title_text:
            
            filtered_titles.append(Title_text)

    #將新聞標題轉換為 DataFrame
    df = pd.DataFrame({category + '即時新聞標題': filtered_titles})

    #重設資料列索引值由 1 開始
    df.index += 1

    # 輸出 DataFrame
    print(df)
    print('\n')
```
