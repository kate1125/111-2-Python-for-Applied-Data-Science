1. **將檔案內所有欄位標題與每一筆資料列印出來（請試著盡可能排列整齊，如下一個 cell 中的輸出示範）；Python3 中如何將中文的輸出對齊，請參考 [[1]](https://www.itread01.com/content/1550554048.html) [[2]](https://www.zhihu.com/question/266249354)**

```python
import csv
from zipfile import ZipFile
from io import TextIOWrapper

from io import BytesIO
import requests  #建立網路連線
import urllib  #去目的地網站上把檔案抓下來
from urllib.request import urlretrieve  #從網路上萃取資料出來

#urlretrieve() 方法用於下載指定的 URL
url = "https://datacenter.taichung.gov.tw/swagger/OpenData/ab6ab38e-de1f-4718-915d-e81ea0761273"  #此為路徑，而非檔案名稱
file_path = urlretrieve(url, "臺中市空氣品質監測站即時監測資料.CSV")  #檔案路徑

with open('臺中市空氣品質監測站即時監測資料.CSV', 'r', newline = '', encoding='utf-8-sig') as csvfile:  #開啟檔案
    
     reader = csv.reader(csvfile)
        
     header_line = next(reader)  #next() 函式會返回『迭代物件』的下一個元素
     header = [f'『{column}』' for column in header_line]  #定義變數header，用for迴圈去抓取欄位標題(有雙引號的)
     headers = [f'{column}' for column in header_line]  #定義變數headers，用for迴圈去抓取欄位標題

     print(f"『{file_path[0]}』共有 {len(header)} 個欄位，分別為：\n {','.join(header)}")
     print()
    
     print("{:<10}\t{:<10}\t{:<15}\t{:<10}\t{:<10}".format(*headers))  #印出資料欄位，運用格式化

     for row in reader:  #for迴圈輸出定義欄位
        
        print("{:<10}\t{:<10}\t{:<15}\t{:<10}\t{:<10}".format(row[0],row[1],row[2],row[3],row[4]))
        
     csvfile.close()
```

### <font color=#0000FF>第二題：</font>
**題目要求完全與第一題一樣，但目標檔案是「臺中市空氣品質監測站即時監測資料.JSON」**
                                                      

```python
import json, requests, urllib  #requests 建立網路連線，urllib 去目的地網站上把檔案抓下來
from io import BytesIO
from urllib.request import urlretrieve  #從網路上萃取資料出來

url = 'https://datacenter.taichung.gov.tw/swagger/OpenData/ad8998ea-a749-40cb-80e5-6e001f262b7b'

urlretrieve(url, "臺中市空氣品質監測站即時監測資料.JSON")  #從網路上下載檔案

with open("臺中市空氣品質監測站即時監測資料.JSON", encoding='utf-8') as jsonfile:  #開啟檔案
    
    data = json.load(jsonfile)
    
    header_line = list(data[0].keys())  #使用list函式取得data中第一個元素的所有鍵名（即欄位名），並將其儲存在header_line變數中
    
    print("『臺中市空氣品質監測站即時監測資料.JSON』", "共有 {} 個欄位".format(len(header_line)), "，分別為：\n", "，".join(["『{}』".format(header) for header in header_line]))
    print()

    print("{:<10}\t{:<10}\t{:<15}\t{:<10}\t{:<10}".format(*header_line))  #印出資料欄位，運用格式化

    for item in data:  #for迴圈列印欄位資料

        print("{:<10}\t{:<10}\t{:<15}\t{:<10}\t{:<10}".format(item['測站名稱'], item['測值時間'], item['監測項目'], item['測值'], item['狀態']))

    jsonfile.close()
```

### <font color=#0000FF>第三題：</font>
**題目要求完全與第一題一樣，但目標檔案是「臺中市空氣品質監測站即時監測資料.XML」**                                                

```python
import xml.etree.ElementTree as et  #匯入解析 XML 文件模組
from io import TextIOWrapper  # I/O 模組，用於將二進位資料轉換為文字格式
from urllib.request import urlretrieve  #下載檔案的模組，用於從網路上下載資料

#urlretrieve() 方法用於下載指定的 URL
url = 'https://datacenter.taichung.gov.tw/swagger/OpenData/072e5e1d-fb39-4bda-847f-f42b090e2921'
filename, headers = urlretrieve(url, "臺中市空氣品質監測站即時監測資料.XML")

tree = et.ElementTree(file=filename)  #ElementTree() 構建空樹
root = tree.getroot()  #獲取根節點，通過下標可訪問相應的節點

print("『", filename, "』","共有", len(root[0]), "個欄位，分別為：\n", '，'.join(["『{}』".format(i.tag) for i in root[0]]))
print()

print("{:<10}\t{:<10}\t{:<15}\t{:<10}\t{:<10}".format(*[i.tag for i in root[0]]))  #印出資料欄位，運用格式化

for i in range(len(root)):  #for迴圈列印欄位資料
    
    print("{:<10}\t{:<10}\t{:<15}\t{:<10}\t{:<10}".format(root[i][0].text,root[i][1].text,root[i][2].text,root[i][3].text,root[i][4].text))
```
